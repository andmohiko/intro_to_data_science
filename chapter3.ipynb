```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
dau_df = pd.read_csv("R/section3-dau.csv")
dau_df.head()
```

```python
dpu_df = pd.read_csv("R/section3-dpu.csv")
dpu_df.head()
```

```python
install_df = pd.read_csv("R/section3-install.csv")
install_df.head()
```

```python
dau_install = pd.merge(dau_df, install_df, on=['user_id', 'app_name'])
dau_install
```

```python
dau_install_payment = pd.merge(dau_install, dpu_df, on=['log_date', 'app_name', 'user_id'], how='outer')
dau_install_payment
```

```python
dau_install_payment = dau_install_payment.fillna(0)
```

```python
#年月日を年月に変更するため、日付のカラムを7桁目までにする
dau_install_payment['log_month'] = dau_install_payment['log_date'].str.slice(0, 7)
dau_install_payment['install_month'] = dau_install_payment['install_date'].str.slice(0, 7)
dau_install_payment
```

```python
mau_payment = dau_install_payment.groupby(['log_month', 'user_id', 'install_month']).sum().reset_index()
mau_payment.head()
```

```python
#install_month と log_month が一致するならinstall, 一致しなければexistingというラベルをつける
mau_payment['user_type'] = np.where(mau_payment['install_month']==mau_payment['log_month'], 'install', 'existiing')
```

```python
mau_payment_summery = mau_payment.groupby(['log_month', 'user_type']).sum().reset_index()
mau_payment_summery
```

```python
mau_payment_summery.pivot('log_month', 'user_type').plot.bar(y='payment', stacked=True)
# mau_payment_summery.plot(kind='bar', stacked=True)
```
